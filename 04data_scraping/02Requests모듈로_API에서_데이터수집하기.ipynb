# 네이버 API로 검색한 결과 최대한 가져오기
* 1회 네이버 서버에 요청
* 응답에서 total을 찾아서 display로 나누어 총 반복할 페이지 산출
* 반복문 작성 및 종료 조건 설정
* 1페이지 결과를 데이터 프레임으로 만들고 리스트에 append
* 반복이 끝난 후 리스트에 append된 데이터프레임을 1개로 합침

```python
import requests
import pandas as pd

url = "https://openapi.naver.com/v1/search/book.json"

def json_unpack(data):
    result = {}
    for item in data['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    df = pd.DataFrame(result)
    return(df)

keyword = input("도서명을 입력하세요: ")

url = "https://openapi.naver.com/v1/search/book.json"
Client_Id = "_U7XmwtmPDpKOYSG5dic"
Client_Secret = "ji5dkOMzd9"
headers = {"X-Naver-Client-Id" : Client_Id, "X-Naver-Client-Secret" : Client_Secret}
payload = {"query" : keyword, "display": 1, "start" : 1, "sort" : 'sim'}
r = requests.get(url, params=payload, headers = headers)
print(r.url)
print(r.status_code)
data = r.json()
json_unpack(data)

print("전체 검색 데이터 수: ", data['total'] ,"검색할 페이지 수 : ", (int(data['total'])//100+1))

```

# While문으로 데이터 모두 가져오기

```python
import requests
import pandas as pd
```

```python
keyword = input("검색키워드를 입력해 주세요: ")
```

```python
def json_unpack(data):
    result = {}
    for item in data['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    df = pd.DataFrame(result)
    return(df)
```

```python
all_data2 = []
start = 1


while True:
    url = "https://openapi.naver.com/v1/search/book.json"
    Client_Id = "_U7XmwtmPDpKOYSG5dic"
    Client_Secret = "ji5dkOMzd9"
    headers = {"X-Naver-Client-Id" : Client_Id, "X-Naver-Client-Secret" : Client_Secret}
    payload = {"query" : keyword, "display": 100, "start" : start, "sort" : 'sim'}
    r = requests.get(url, params=payload, headers = headers)
    print(r.url)
    print(r.status_code)
    data = r.json()
    ######### 여기까지가 1바퀴 ##########


    n_data = data['total']
    total_pages = (int(data['total'])//100)+1
    print("전체 검색 데이터 수: ", n_data,"검색할 페이지 수 : ", total_pages)
    result_df = json_unpack(data)
    all_data2.append(result_df)
    if start == 1000:
        break 
    elif start < 901:
        start += 100
    elif start == 901:
        start += 99
    
```

```python
len(all_data2)
```

```python
final_result2 = pd.concat(all_data2)
final_result2 = final_result2.reset_index(drop=True)
final_result2
```

```python
final_result2.to_csv("네이버검색_책_파이썬.csv", encoding="utf-8-sig", index=False)
```

# 뉴스, 블로그, 책 한번에 검색해서 저장하기

```python
all_data3 = []

for category in ["blog", "news", "book"]:
    start = 1
    while True:
        url = f"https://openapi.naver.com/v1/search/{category}.json"
        Client_Id = "_U7XmwtmPDpKOYSG5dic"
        Client_Secret = "ji5dkOMzd9"
        headers = {"X-Naver-Client-Id" : Client_Id, "X-Naver-Client-Secret" : Client_Secret}
        payload = {"query" : keyword, "display": 100, "start" : start, "sort" : 'sim'}
        r = requests.get(url, params=payload, headers = headers)
        print(r.url)
        print(r.status_code)
        data = r.json()
        ######### 여기까지가 1바퀴 ##########


        n_data = data['total']
        total_pages = (int(data['total'])//100)+1
        print("전체 검색 데이터 수: ", n_data,"검색할 페이지 수 : ", total_pages)
        result_df = json_unpack(data)
        all_data3.append(result_df)
        if start == 1000:
            break 
        elif start < 901:
            start += 100
        elif start == 901:
            start += 99

final_result3 = pd.concat(all_data3)
final_result3 = final_result3.reset_index(drop=True)
final_result3
```
