# 3. requests 모듈 사용법
**1) 모듈 불러오기: import requests** <br>
**2) url 변수에 저장** <br>
**3) url 파라미터** <br>
    * payload = {파라미터1: value, 파라미터2: value} <br>
**4) headers** <br>
    * headers = {key:value} <br>
**5) HTTP 요청 보내기 get, post** <br>
    * get: r = requests.get(url, params=payload, headers=headers)<br>
    * post: r = requests.post(url, data={key:value})<br>
**6) 만들어진 url 확인: print(r.url)** <br>
**7) 응답 코드 확인: print(r.status_code)** <br>
**8) 응답 요소 출력: r.text, r.json(), r.content** <br>

```python
def json_unpack(all_data):
    final_result = {}
    for page in all_data:
        for item in page['items']:
            for key, value in item.items():
                final_result.setdefault(key, []).append(value)
    df = pd.DataFrame(final_result)
    return(df)
```

* 응답에서 total을 찾아서 display로 나누어 총 반복할 페이지 산출
* 반복문 작성 및 종료 조건 설정
* 1페이지 결과를 데이터 프레임으로 만들고 리스트에 append
* 반복이 끝난 후 리스트에 append된 데이터프레임을 1개로 합침

```python
import requests
import pandas as pd

url = "https://openapi.naver.com/v1/search/book.json"

def json_unpack(data):
    result = {}
    for item in data['items']:
        for key, value in item.items():
            result.setdefault(key, []).append(value)
    df = pd.DataFrame(result)
    return(df)

keyword = input("도서명을 입력하세요: ")

url = "https://openapi.naver.com/v1/search/book.json"
Client_Id = "_U7XmwtmPDpKOYSG5dic"
Client_Secret = "ji5dkOMzd9"
headers = {"X-Naver-Client-Id" : Client_Id, "X-Naver-Client-Secret" : Client_Secret}
payload = {"query" : keyword, "display": 1, "start" : 1, "sort" : 'sim'}
r = requests.get(url, params=payload, headers = headers)
print(r.url)
print(r.status_code)
data = r.json()
json_unpack(data)

print("전체 검색 데이터 수: ", data['total'] ,"검색할 페이지 수 : ", (int(data['total'])//100+1))

```
