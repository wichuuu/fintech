# matplotlib의 2가지 코딩 스타일 

## 00 - 스타일: matplotlib 공식 추천 스타일
* 복잡한 그래프를 그릴 때 사용
* 사용법이 pyplot 스타일보다 복잡함

```python
# 0부터 2까지를 100개의 점으로 나눈 배열

x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots(figsize = (5, 2.7))
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label = "quadratic")
ax.plot(x, x**3, label= "cubic")
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(5, 2.7))
print(ax)
ax[0].plot(x, x, label = 'linear')
ax[1].plot(x, x**2, label = 'quadratic')
ax[2].plot(x, x**3, label = 'cubic')

ax[0].set_xlabel('x_label')
ax[0].set_ylabel('y_label')
ax[1].set_title('간단 그래프')

ax[1].set_xlabel('x_label')
ax[1].set_ylabel('y_label')
ax[1].set_title('간단 그래프')

ax[2].set_xlabel('x_label')
ax[2].set_ylabel('y_label')
ax[2].set_title('간단 그래프')

ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

## pyplot 스타일
* OO-style 보다 사용법 간단
* 복잡한 그래프를 그릴 때는 적합하지 않다

```python
plt.figure(figsize=(5, 2.7))
plt.plot(x, x, label='linear') # 직선
plt.plot(x, x**2, label='quadratic') # y = x ^ 2, 기울기가 더 큰 직선
plt.plot(x, x**3, label='cubic') # y = x ^ 3, 곡선 
plt.xlabel("X축 이름")
plt.ylabel("Y축 이름")
plt.title("pyplot 스타일 그래프")
plt.legend()
plt.show
```

# 선 모양 바꾸기 set_linestyle(모양)

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python
data1
```

### 선 모양: '-' , '--', '-.', ':', ''

```python
len(data1)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle=":")
ax.plot(x, np.cumsum(data2), color='orange', linewidth=2, linestyle="-.")
plt.show()
```

# 산점도 그래프를 그리고 색 변경하기
* facecolor(안쪽색), edgecolor(테두리색)

컬러를 RGB 값으로 주고 싶으면 (0, 1, 0.5) 와 같이 0-1 사이 값을 주면 됨

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s=50, facecolor='C5', edgecolor='k')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s= 50, facecolor=(0.5, 0.5, 0.5), edgecolor='k')
plt.show()
```

# 마커의 모양 지정하기


| 마커 | 설명 | |-----|:-----| |'.' | point marker | |',' | pixel marker | |'o' | circle marker| |'v' | triangle_down marker| |'^' | triangle_up marker| |'<' | triangle_left marker| |'>' | triangle_right marker| |'1' | tri_down marker| |'2' | tri_up marker| |'3' | tri_left marker| |'4' | tri_right marker| |'8' | octagon marker| |'s' | square marker| |'p' | pentagon marker| |'P' | plus (filled) marker| |'*' | star marker| |'h' | hexagon1 marker| |'H' | hexagon2 marker| |'+' | plus marker| |'x' | x marker| |'X' | x (filled) marker| |'D' | diamond marker| |'d' | thin_diamond marker| |'|' | vline marker| |'_' | hline marker|

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(data1, '8', label='data1')
ax.plot(data2, '2', label='data2')
ax.plot(data3, 'v', label='data3')
ax.plot(data4, 's', label='data4')
ax.legend()
plt.show()
```

# X축, Y축 이름, 눈금, 그래프이름

```python
#alpha 값은 투명도

mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
n, bins, patches = ax.hist(x, 50, density = True, facecolor = 'C0', alpha=0.75)
ax.set_xlabel("Length [cm]")
ax.set_ylabel("Probability")
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115, \ \sigma=15$')
ax.axis([60, 160, 0, 0.03]) #x, y축의 범위
# ax.axis([])
plt.show()
```

# 눈금자 단위 구간 설정하기

```python
xdata = np.arange(len(data1))

fig, ax = plt.subplots(2, 1,figsize = (5, 2.7), layout = 'constrained')
ax[0].plot(xdata, data1)
ax[0].set_title('Automatic ticks')

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title("Manual ticks")
plt.show()
```
