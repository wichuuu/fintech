# 리스트 표현식
* 리스트 안에 for 반복문과 if else 조건문을 한 번에 적어서 리스트를 만드는 방법
* 여러 줄의 코드를 1줄로 줄일 수 있다.
* for 문과 else문을 사용해서 처리할 때보다 속도가 빠르다.
* [실행할 명령 for 변수 in list, tuple]

```python
l1 = []
for i in range(1, 11):
    l1.append(i)
l1
```

```python
[i for i in range(1, 11)]
```

```python
import time
```

```python
start = time.time()
a = []
for i in range(1, 200000000):
    a.append(i + 5)
print(a[:10])

print("코드 실행 시간: ", time.time() - start)
```

```python
start2 = time.time()
b = [i+5 for i in range(1, 200000000)]
print(b[:10])
print("코드 실행 시간: ", time.time() - start2)
```

* 1에서 10까지의 숫자 중 짝수만 더해서 리스트 만들기

```python
l1 = []
for i in range(1, 11):
    if i % 2 == 0:
        l1.append(i)
l1
```

```python
l2 = [i for i in range(1, 11) if i % 2 == 0]
l2
```

## 다중 반복문을 사용한 리스트 표현식
[i * j for i in range(2, 10) for j in range(1, 9)]

```python
gugu = []
for i in range(2, 10):
    for j in range(1, 10):
        gugu.append(i * j)  # gugu.append(f"{i} x {j} = {i*j}"cv)
print(gugu)
```

## with open

```python
with open("test2.txt", 'w') as file3:
    file3.write("with open을 사용하면 파일을 자동으로 닫아주어 편리하다.")
```

```python
with open("test2.txt", 'r') as file4:
    data = file4.read()
print(data)
```

# encoding에 따른 읽기 오류
* 운영체제나 (windows, linux, macOS), 한글/영어 윈도우에 따라서 인코딩 되는 포멧이 달라짐
* 인코딩에 따른 오류가 발생 가능
* web 표준은 utf-8 이므로 인코딩은 utf-8로 하는 게 좋음
* windows 경우에는 인코딩이 cp949 이므로 주의

```python
with open("test_utf-8.txt", 'w', encoding='utf-8') as file5:
    file5.write("이것은 with open으로 만들 utf-8 인코딩 파일입니다.")
```

```python
with open("test_utf-8.txt", 'r', encoding="utf-8") as file6:
    file6_data = file6.read()
print(file6_data)
```
