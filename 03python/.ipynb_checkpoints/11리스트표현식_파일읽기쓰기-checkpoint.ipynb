# 리스트 표현식
* 리스트 안에 for 반복문과 if else 조건문을 한 번에 적어서 리스트를 만드는 방법
* 여러 줄의 코드를 1줄로 줄일 수 있다.
* for 문과 else문을 사용해서 처리할 때보다 속도가 빠르다.
* [실행할 명령 for 변수 in list, tuple]

```python
l1 = []
for i in range(1, 11):
    l1.append(i)
l1
```

```python
[i for i in range(1, 11)]
```

```python
import time
```

```python
start = time.time()
a = []
for i in range(1, 200000000):
    a.append(i + 5)
print(a[:10])

print("코드 실행 시간: ", time.time() - start)
```

```python
start2 = time.time()
b = [i+5 for i in range(1, 200000000)]
print(b[:10])
print("코드 실행 시간: ", time.time() - start2)
```

* 1에서 10까지의 숫자 중 짝수만 더해서 리스트 만들기

```python
l1 = []
for i in range(1, 11):
    if i % 2 == 0:
        l1.append(i)
l1
```

```python
l2 = [i for i in range(1, 11) if i % 2 == 0]
l2
```

## 다중 반복문을 사용한 리스트 표현식
[i * j for i in range(2, 10) for j in range(1, 9)]

```python
gugu = []
for i in range(2, 10):
    for j in range(1, 10):
        gugu.append(i * j)  # gugu.append(f"{i} x {j} = {i*j}"cv)
print(gugu)
```

```python
[f"{i} x {j} = {i*j}" for i in range(2, 10) for j in range(1, 10)]
```

# 파일 읽기 쓰기 open(), with open()
## 1) 파일에 문자열을 쓰거나 읽을 땐 open() 함수 사용
* 파일 저장
변수명 = open("파일 이름", 모드(w:쓰기, r:읽기)<br>
변수명.write("저장할 내용")<br>
변수명.close()<br>
**close()를 안 하면 계속 열려 있음** 

```python
file = open("test.txt", "w")
file.write("파일 만들기 연습")
file.close()
```

* 파일열기
변수명1 = open("파일 이름", 파일 모드(r)) <br>
변수명2 = 변수명1.read() <br>
변수명1.close() <br>

```python
file1 = open("test.txt", "r")
my_file = file1.read()
file1.close()
print(my_file)
```

## with open
* with open은 close()를 하지 않아도 자동으로 close() 처리를 해줌

```python
with open("test2.txt", 'w') as file3:
    file3.write("with open을 사용하면 파일을 자동으로 닫아주어 편리하다.")
```

```python
with open("test2.txt", 'r') as file4:
    data = file4.read()
print(data)
```

# encoding에 따른 읽기 오류
* 운영체제나 (windows, linux, macOS), 한글/영어 윈도우에 따라서 인코딩 되는 포멧이 달라짐
* 인코딩에 따른 오류가 발생 가능
* web 표준은 utf-8 이므로 인코딩은 utf-8로 하는 게 좋음
* windows 경우에는 인코딩이 cp949 이므로 주의

```python
with open("test_utf-8.txt", 'w', encoding='utf-8') as file5:
    file5.write("이것은 with open으로 만들 utf-8 인코딩 파일입니다.")
```

```python
with open("test_utf-8.txt", 'r', encoding="utf-8") as file6:
    file6_data = file6.read()
print(file6_data)
```

# json 파일 읽어와서 내용 정리하기

```python
with open("./data/서울특별시_관광지입장정보_2011_2016.json", 'r', encoding="utf-8") as file7:
    trip_info = file7.read()
trip_info
```

```python
print(type(trip_info))
```

```python
import json 
```

```python
data = json.loads(trip_info)
```

```python
data
```

```python
data[:2]
```

```python
for item in data[0]: #딕셔너리를 반복문에 넣으면 key가 나옴
    print(item)
```

```python
for item in data[2].items(): 
    print(item)
```

```python
for item in data[:2]:
    print(item['ForNum'])
    print(item['NatNum'])
    print(item['addrCd'])
    print(item['gungu'])
    print(item['resNm'])
    print(item['sido'])
    print(item['yyyymm'])
```

```python
year_mon = []
sido = []
place = []
foriners = []
total = []

for item in data:
    foriners.append(item['ForNum'])
    total.append(item['NatNum'])
    place.append(item['resNm'])
    sido.append(item['sido'])
    year_mon.append(item['yyyymm'])
    
tour_info = dict(년월=year_mon, 시도=sido, 장소=place, 외국인=foriners, 전체방문객수=total)
tour_info
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tour_info)
df
```

* value를 리스트로 만든 후 dict 함수로 딕셔너리로 만들어 데이터프레임 만들기

* 딕셔너리를 바로 만들어서 데이터프레임으로 변환
* key, value를 직접 할당해서 딕셔너리 만들기

```python
d = {}
d['장소'] = "경복궁"
d
```

```python
result = {}
for key in data[0]: 
    value_list = []
    for item in data:
        value_list.append(item[key])
    print(key, value_list)
    result[key] = value_list
    
result
```

```python
result["NatNum"]
```

```python
result_df = pd.DataFrame(result)
result_df
```

```python
d2 = {}
d2.setdefault("key", 3)
d2
```

**setdefalut 함수를 이용해 만들기**

```python
dic_result = {}
for dicts in data:
    for key, value in dicts.items():
#         print(key, value)
        dic_result.setdefault(key, []).append(value)
```

```python
dic_result_df = pd.DataFrame(dic_result)
dic_result_df
```
