```python
import requests
import json
import time
import pandas as pd

url = "https://apihub.kma.go.kr/api/typ01/url/rdr_stn_file_list.php"

params = {
    "stn": "KWK",               # 관악산 관측소 코드
    "rdr": "HSR",               # HSR: 수상체
    "tm": "20210815",           # 날짜 (형식: YYYYMMDD)
    "size": "Y",                # 파일 크기 포함 여부
    "authKey": "0RI5yfi7QceSOcn4u8HHlA"  # 발급받은 인증키
}

response = requests.get(url, params=params)

try:
    data = response.json()
    print(json.dumps(data, indent=2, ensure_ascii=False))
except Exception as e:
    print("응답 오류:", e)
    print("텍스트 응답:", response.text)
```
