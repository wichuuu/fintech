```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs

stock_infos = []
page = 1

while True:
    url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
    payload = dict(
        method="searchCorpList",
        pageIndex=page, 
        currentPageSize=100,
        orderMode=3,
        orderStat="D",
        searchType=13,
        fiscalYearEnd="all",
        location="all"
    )

    r = requests.post(url, data=payload)
    soup = bs(r.content, 'lxml')

    
    total_page = int(soup.select_one(".info.type-00 > em").text.replace(",", ""))

    stock = []

    for idx, tr in enumerate(soup.select("tbody > tr")):
        stock_type = tr.select_one("td:nth-child(1) > img")['alt']
        company = tr.select_one("td:nth-child(1) > a")["title"]
        code = tr.select_one("td:nth-child(1) > a")["onclick"].split("'")[1]
        company_type = tr.select_one("td:nth-child(2)").text
        product = tr.select_one("td:nth-child(3)").text
        date = tr.select_one("td:nth-child(4)").text
        month = tr.select_one("td:nth-child(5)").text
        ceo = tr.select_one("td:nth-child(6)").text
        hompage = tr.select_one("td:nth-child(7) > a")["href"] if tr.select_one("td:nth-child(7) > a") else ""
        region = tr.select_one("td:nth-child(8)").text

        stock.append((
            stock_type, company, code, company_type,
            product, date, month, ceo, hompage, region
        ))

    
    stock_infos.extend(stock)

    if page * 100 >= total_page:
        break
    else:
        page += 1


columns = soup.select_one("table")["summary"].split(", ")
columns.insert(0, "증권종류")
columns.insert(2, "종목코드")

df = pd.DataFrame(stock_infos, columns=columns)
print(df.head())
```
